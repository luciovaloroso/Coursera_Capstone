Now it's time to add some geospatial data to dataset. We first download geojson file and then extract some information relative to each zone.

```python
!wget --quiet https://capstone-prj-re-ita.s3.us-east-2.amazonaws.com/H501.geojson -O roma_zones.json

print('GeoJSON file downloaded!')
```

Geojson contains the points of the polygons describing each zone. <br/>It would be fine to get the approximate center of each polygon in order to explore each zone with more precision with foursquare apis.

```python
!conda install shapely
```

```python
import json
from shapely.geometry import shape
```

```python
zones['lat'] = 0.0
zones['lon'] = 0.0

with open('roma_zones.json') as roma_geo:
    json_data = json.load(roma_geo)
       
    for p in json_data['features']:
        cod_zona = p['properties']['CODZONA']
        #print('Zona: ' + cod_zona)
        centroid_lat = shape(p['geometry']).buffer(0).centroid.y
        centroid_lon = shape(p['geometry']).buffer(0).centroid.x
        #print('Centroid: ' + str(shape(p['geometry']).buffer(0).centroid))
        zones.loc[zones['Zona']==cod_zona,'lat'] = centroid_lat
        zones.loc[zones['Zona']==cod_zona,'lon'] = centroid_lon

zones.head()
```

Let's restrict our analysis to residential houses in normal state of maintenance. Since we have minimum and maximum price for each zone it could be useful to add a column with average price.

```python
prices = prices[prices['Stato']=='NORMALE']
prices = prices[prices['Descr_Tipologia']=='Abitazioni civili']
prices['Compr_avg'] = (prices['Compr_max']+prices['Compr_min'])/2
prices.head()
```

Merge the two dataframes keeping only useful columns.

```python
roma_prices = pd.merge(prices, zones[['Zona','Zona_Descr','lat','lon']], on='Zona')
```

Print a map of Rome

```python
roma_lat=41.890132
roma_lon=12.492483

roma_map = folium.Map(location=[roma_lat, roma_lon], zoom_start=11)

#display map
roma_map
```

```python
roma_geo = r'roma_zones.json' # geojson file

# generate choropleth map using the average price for each zone
roma_map.choropleth(
    geo_data=roma_geo,
    data=roma_prices,
    columns=['Zona', 'Compr_avg'],
    key_on='feature.properties.CODZONA',
    fill_color='YlOrRd', 
    fill_opacity=0.7, 
    line_opacity=0.2,
    legend_name='House prices'
)

# display map
roma_map
```
